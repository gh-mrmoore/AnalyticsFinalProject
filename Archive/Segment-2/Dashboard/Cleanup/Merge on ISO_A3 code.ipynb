```python
# Add the Pandas dependency.
import pandas as pd
```

```python
victims = 'Victims_iso_a2.csv'
countries = 'Country_coordinates.csv'
```

```python
df1 = pd.read_csv(victims)
df2 = pd.read_csv(countries)
```

```python
df1.head()
```

```python
df3 = pd.merge(df1, df2, on='iso_a2')
df3.head()
```

```python
df4=df3.rename(columns={'iso_a3_y': 'iso_a3'})
df4.head()
```

```python
df5 = df4.drop(columns=['iso_a3_x','Unnamed: 0', 'latitude', 'longitude'])
df5
```

```python
df5.to_csv('Victims_per_country.csv', index=False)
```

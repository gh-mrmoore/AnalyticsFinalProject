```python
# Dependencies
import os
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from config import conn
```

```python
# Create engine with connection string
engine = create_engine(conn)

# Get saved WEO unemployment data and convert to DataFrame
unemployment_df = pd.read_csv("https://raw.githubusercontent.com/gh-mrmoore/AnalyticsFinalProject/MattMoore/Data/WEO_Unemployment.csv")
unemployment_df
```

```python
# Export DataFrame to PostgreSQL table
unemployment_df.to_sql('WEO_Unemployment', con=engine, if_exists='append', index=False)
```
